#### From the Fitbit data, we want to visually understand:

```python
months = fitbit.groupby("Months_encoded")['Steps'].median()
w_w = fitbit.groupby("Work_or_Weekend")['Steps'].median()
w_w
```

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here
fig, [ax1, ax2] = plt.subplots(nrows=1, ncols=2, figsize=(12,4))
plt.sca(ax1)
months.plot()
plt.xlabel('Month', fontsize=12)
plt.ylabel('Median Steps', fontsize=12)

# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
plt.sca(ax2)
w_w.plot(kind='bar')
plt.xticks(range(2), ["Weekend", "Workdays"])
plt.xlabel('')
plt.ylabel('Median Steps', fontsize=12)

plt.savefig('Median_steps_per_month.png', dpi = 200, quality=100)
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here
fig, ax = plt.subplots(3,1,figsize=(17,18))
features = ['Minutes Lightly Active','Minutes Very Active','Minutes Sedentary']
for i,j in zip(features,ax):
    j.scatter(fitbit[i],fitbit['Steps'], s=50, color='blue', edgecolor='black')
    j.set_xlabel(i)
    j.set_ylabel('Steps')
plt.show()
```
